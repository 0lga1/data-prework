### 1st Data Source - Happiness report

```python
# Import 1st Data Source
happiness_data = pd.read_csv('2019.csv')
happiness_data.head()
```

```python
happiness_data = happiness_data.rename(columns={'Country or region': 'country'})
```

```python
# Get list of all countries in dataset
input_countries = happiness_data['country'].tolist()
print(input_countries)
```

```python
# Conversion to ISO3166 code, will be needed later 
import pycountry

countries = {}
for country in pycountry.countries:
    countries[country.name] = country.alpha_2

codes = [countries.get(country, 'Unknown code') for country in input_countries]

print(codes) 
```

```python
#Add new column to the table with country ISO3166 code
happiness_data.insert(2, "Country_code", codes, True) 
happiness_data.head()
```

```python
codes_cleaned = [e for e in codes if e != 'Unknown code']
print(codes_cleaned)
```

### 2nd Data Source - Holidays API

```python
#Retrieve list of holidays 2019 for hall countries

holidays_list=pd.DataFrame()

countries = codes_cleaned
for country in countries:
    k = country
    url_holiday =f'https://calendarific.com/api/v2/holidays?&api_key=0e31f467e23a99052f69cbbd14fc74589081d3b6&country={k}&year=2019/json'
    response = requests.get(url_holiday, headers ={"Content-Type": "application/json", "Accept": "application/json", "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36"})
    results = response.json()
    flattened_data = pd.json_normalize(results['response']['holidays'])
    flattened_data
    holidays_list = holidays_list.append(flattened_data)

#print and review the data you obtained
holidays_list.
```

```python
holidays_list['country.name'].value_counts()
```

```python
#What type of holidays the dataset contains?
holidays_list['type'].value_counts()
```

```python
#Select only public (national holidays)
public_holidays = holidays_list[ holidays_list['type'].astype(str).str.contains('National') ]
public_holidays.head()
```

```python
#How many public holidays there were in 2019 in each country?
#Which country had the most and the least public holidays in 2019?
public_holidays['country.name'].value_counts()
```

```python
#The most celebrated 
print(public_holidays['name'].value_counts().index[0])
print(public_holidays['name'].value_counts().max())
```

```python
#Which country had the most and the least religious holidays in 2019?
religious_holidays = holidays_list[ holidays_list['type'].astype(str).str.contains('Christian|Orthodox|Hebrew|Muslim|Hinduism') ]
religious_holidays

religious_holidays['country.name'].value_counts().head()
```

```python
#create new df with nb of holidays per country
public = pd.DataFrame(public_holidays['country.name'].value_counts().reset_index())
public.columns = ['country', 'count_public']
public.head()

religious = pd.DataFrame(religious_holidays['country.name'].value_counts().reset_index())
religious.columns = ['country', 'count_religious']
religious.head()
```

```python
# merge count religious and public holidays per country
merged = pd.merge(public, religious, how='outer', on='country')
merged.head()
```

```python
#add column total nb of holidays
merged['sum'] = (merged['count_public'].fillna(0) + merged['count_religious'].fillna(0)).astype(int)
merged
```

```python
county_happy_holiday = pd.merge(happiness_data, merged, how='outer', on='country')
county_happy_holiday.head()
```

```python
#Is there is correlation between nb holidays and hqppiness score?
county_happy_holiday['Score'].corr(county_happy_holiday['sum'])
```

```python
#Visualisation
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')

plt.scatter(county_happy_holiday['Score'], county_happy_holiday['sum'])
plt.show()
```
